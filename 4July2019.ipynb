```python
# Linear graph

import matplotlib.pyplot as plt

plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph to plot distance and time

distance=list(range(2,11,2))
time=list(range(1,6,1))

# d=2*t
plt.title('Distance vs Time')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(time,distance,'bo')
plt.show()
```

```python
# V= at
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]

plt.plot(t,v)
plt.grid()
plt.show()
```

```python
n=list(range(10))
sqn=[i**2 for i in n]

plt.plot(n,sqn)
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.title('Numbers Vs Squares')
plt.show()
```

```python
# Voltage - Time
import numpy as np

t=[i/10 for i in range(1,21)]

v=[1+np.sin(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title('Voltage Vs Time Graph')
plt.savefig('Images/Voltage-Time.png')
plt.show()
```

```python
# Plot Income of Arkansas from 2005 to 2013

import pandas as pd
df=pd.read_csv('DataFiles/income.csv')
years=df.columns[2:]

incomeArkansas=df.values[3][2:]
plt.title('Income of Arkansas')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,incomeArkansas)
plt.show()
```

```python
# Income of all states from 2005 to 2013

incomeallstates=[]

for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
    
plt.plot(years,incomeallstates)
plt.title('All states Income')
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()
incomeallstates
```

```python
def highestrem(n):
    dict={}
    for i in range(1,n):
        rem=n%i
        dict[i]=rem
    maxrem=max(dict.values())
    for item in dict.items():
        if item[1]==maxrem:
            print(item[0])
n=int(input())
highestrem(n)
```
